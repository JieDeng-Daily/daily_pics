```python
import pandas as pd
import numpy as np
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.colors as mcolors
```

```python
def color_structure(database,vmin,vmax,name):
    database = database.set_index('Number').T
    values = database.loc['Mutaionrate']   
    red_white_cmap = LinearSegmentedColormap.from_list('WhiteRed', [ '#FFFFFF','#FF0000'])
    #norm = mcolors.Normalize(vmin=min(values), vmax=max(values))
    norm= mcolors.Normalize(vmin, vmax)
    cmap = red_white_cmap
        
    rgb_colors = {key: mcolors.rgb2hex(cmap(norm(value))) for key, value in database.loc['Mutaionrate'].items()}
    print('color #1 white;lighting soft;graphics silhouettes True')
    for key, color in rgb_colors.items():
        print(f"color #1: {key} {color}")
    print(f"view for;save {name}{vmin}-{vmax}_for.png transparentbackground True")
    print(f"view top;save {name}{vmin}-{vmax}_top.png transparentbackground True")

def color_structure_v2(database,vmin,vmax,name,list):
    sel_database = database[database['Number'].isin(list)]
    sel_database = sel_database.set_index('Number').T
    values = sel_database.loc['Mutaionrate']   
    red_white_cmap = LinearSegmentedColormap.from_list('WhiteRed', [ '#FFFFFF','#FF0000'])
    #norm = mcolors.Normalize(vmin=min(values), vmax=max(values))
    norm= mcolors.Normalize(vmin, vmax)
    cmap = red_white_cmap
        
    rgb_colors = {key: mcolors.rgb2hex(cmap(norm(value))) for key, value in sel_database.loc['Mutaionrate'].items()}
    print('color #1 white;lighting soft;graphics silhouettes True')
    for key, color in rgb_colors.items():
        print(f"color #1: {key} {color}")
    print(f"view for;save {name}-{vmin}-{vmax}_for.png transparentbackground True")
    print(f"view top;save {name}-{vmin}-{vmax}_top.png transparentbackground True")

def generate_flatpng_forline(name,epilist):
    epi = str(epilist).strip('[]')
    print(f"color #1 white;lighting flat;graphics silhouettes false")
    print(f"color #1:{epi} red")
    print(f"view for;save {name}_for.png transparentbackground True")
    print(f"view top;save {name}_top.png transparentbackground True")
```

```python
#database from WJY if need color by mutation rate
rsv_a = pd.read_csv('b_realmut/rsva_all.use.ref.aln.line.matrix.fasta.statistics.txt',sep = '\t')
rsv_b = pd.read_csv('b_realmut/rsvb_all.use.ref.aln.line.matrix.fasta.statistics.txt',sep = '\t')
```

```python
#overall
vmin = 0
vmax = 0.1
color_structure(rsv_a,vmin,vmax,'rsv_a')
color_structure(rsv_b,vmin,vmax,'rsv_b')
```

```python
epi = pd.read_csv('./Epi_4.csv',header=None,index_col=0)
epi['epi_list'] = epi.apply(lambda row: [int(x) for x in row if pd.notna(x)], axis=1)
epi_dict = dict(zip(epi.index,epi['epi_list']))
```

```python
#merge
epi_all = []
for key,value in epi_dict.items():
    epi_all+=value
epi_all = list(set(epi_all))
vmin = 0
vmax = 0.1
color_structure_v2(rsv_a,vmin,vmax,f"rsv_a_merge",epi_all)
color_structure_v2(rsv_b,vmin,vmax,f"rsv_b_merge",epi_all)
```

```python
#each_ab
for key,value in epi_dict.items():
    #print(value)
    vmin = 0
    vmax = 0.1
    color_structure_v2(rsv_a,vmin,vmax,f"rsv_a_{key}",value)
    color_structure_v2(rsv_b,vmin,vmax,f"rsv_b_{key}",value)
```

```python
#forLine
for key,value in epi_dict.items():
    generate_flatpng_forline(key,value)
```
